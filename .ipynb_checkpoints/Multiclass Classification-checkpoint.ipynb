```python
import numpy as np
import pandas as pd
from pandas import Series , DataFrame

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

from sklearn import linear_model
from sklearn.datasets import load_iris

```

```python
# Import the data
iris = load_iris()

# Grab features (X) and the Target (Y)
X = iris.data

Y = iris.target

# Show the Built-in Data Description
print iris.DESCR
```

```python
iris_data = DataFrame(X,columns=['Sepal Length','Sepal Width','Petal Length','Petal Width'])

iris_target = DataFrame(Y,columns=['Species'])
```

```python
iris_data.head()
```

```python
def flower(species):
    if species == 0:
        return 'Setosa'
    elif species == 1:
        return 'Veriscolour'
    else:
        return 'Virginica'
    
iris_target['Species'] = iris_target['Species'].apply(flower)
```

```python
iris_target.head()
```

```python
iris_target.head()
```

```python
iris = pd.concat([iris_data,iris_target],axis=1)
```

```python
iris.head()
```

```python
sns.pairplot(iris,hue='Species',size=2)
```

```python
sns.countplot(x='Petal Length',data=iris,hue='Species')
```

```python
sns.countplot(x='Petal Width',data=iris,hue='Species')
```

```python
# Import SciKit Learn Log Reg
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
logreg = LogisticRegression()
```

```python
# Split the data into Trainging and Testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5,random_state=3)
```

```python
logreg.fit(X_train,Y_train)
```

```python
# Import testing metrics from SciKit Learn
from sklearn import metrics

# Prediction from X_test
Y_pred = logreg.predict(X_test)

#Check accuracy
print metrics.accuracy_score(Y_test,Y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(X_train,Y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
print metrics.accuracy_score(y_pred,Y_test)
```

```python
k_range = range(1,21)

accuracy = []
```

```python
for k in k_range:
    
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,Y_train)
    Y_pred = knn.predict(X_test)
    
    accuracy.append(metrics.accuracy_score(Y_test,Y_pred))
```

```python
accuracy
```

```python
plt.plot(k_range,accuracy)
plt.xlabel('K value')
plt.ylabel('Testing Accuracy')
```
